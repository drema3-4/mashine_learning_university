# Обработка пустых значений

Посмотрим на количество пустых значений

```python
data.info()
```

Восстановление категориальных столбцов, скорее всего, не имеет смысла, так как пропущенных значений очень мало + восстановить нормально не особо получается

Удаление строк с пустыми категориальными столбцами

```python
cat_cols = phob + ['Gender', 'Left - right handed', 'Education', 'Only child', 'Village - town', 'House - block of flats']
```

```python
data = data.dropna(subset=cat_cols)
```

Пропуски в возрасте тоже отборосим

```python
data = data.dropna(subset=['Age'])
```

Пропуски в росте и весе заполлним средними по полу

```python
data.loc[(data['Weight'].isna()) & (data['Gender'] == 'male'), ['Weight']] = \
    data.loc[(data['Weight'].isna()) & (data['Gender'] == 'male'), ['Weight']].fillna(data[(data['Gender'] == 'male')]['Weight'].mean())

data.loc[(data['Weight'].isna()) & (data['Gender'] == 'female'), ['Weight']] = \
    data.loc[(data['Weight'].isna()) & (data['Gender'] == 'female'), ['Weight']].fillna(data[(data['Gender'] == 'female')]['Weight'].mean())
```

```python
data.loc[(data['Height'].isna()) & (data['Gender'] == 'male'), ['Height']] = \
    data.loc[(data['Height'].isna()) & (data['Gender'] == 'male'), ['Height']].fillna(data[(data['Gender'] == 'male')]['Height'].mean())

data.loc[(data['Height'].isna()) & (data['Gender'] == 'female'), ['Height']] = \
    data.loc[(data['Height'].isna()) & (data['Gender'] == 'female'), ['Height']].fillna(data[(data['Gender'] == 'female')]['Height'].mean())
```

Пропуски в количестве родственником заполним средним

```python
data.loc[data['Number of siblings'].isna(), ['Number of siblings']] = \
    data.loc[data['Number of siblings'].isna(), ['Number of siblings']].fillna(int(data['Number of siblings'].mean()))
```

**Таким образом сделаем два новых датасета в одном будем использовать OrdinalEncoder, во втором - OneHotEncoder**

**OrderEncoder:**

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
ord_data = data.copy()

ordEnc = OrdinalEncoder(categories = [['female', 'male']])
ord_data[['Gender']] = ordEnc.fit_transform(ord_data[['Gender']])

ordEnc = OrdinalEncoder(categories = [['left handed', 'right handed']])
ord_data[['Left - right handed']] = ordEnc.fit_transform(ord_data[['Left - right handed']])

ordEnc = OrdinalEncoder(categories = [['college/bachelor degree', 'secondary school', 'primary school', 'masters degree', 'doctorate degree', 'currently a primary school pupil']])
ord_data[['Education']] = ordEnc.fit_transform(ord_data[['Education']])

ordEnc = OrdinalEncoder(categories = [['no', 'yes']])
ord_data[['Only child']] = ordEnc.fit_transform(ord_data[['Only child']])

ordEnc = OrdinalEncoder(categories = [['village', 'city']])
ord_data[['Village - town']] = ordEnc.fit_transform(ord_data[['Village - town']])

ordEnc = OrdinalEncoder(categories = [['block of flats', 'house/bungalow']])
ord_data[['House - block of flats']] = ordEnc.fit_transform(ord_data[['House - block of flats']])
```

**OneHotEncoder:**

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import LabelEncoder
```

```python
oneHotEnc = OneHotEncoder()
dummy_data = data.copy()

res = oneHotEnc.fit_transform(dummy_data[['Gender']])
dummy_data[oneHotEnc.categories_[0]] = res.toarray()
dummy_data.drop('Gender', axis=1, inplace=True)

res = oneHotEnc.fit_transform(dummy_data[['Left - right handed']])
dummy_data[oneHotEnc.categories_[0]] = res.toarray()
dummy_data.drop('Left - right handed', axis=1, inplace=True)

ordEnc = OrdinalEncoder(categories = [['college/bachelor degree', 'secondary school', 'primary school', 'masters degree', 'doctorate degree', 'currently a primary school pupil']])
dummy_data[['Education']] = ordEnc.fit_transform(dummy_data[['Education']])

res = oneHotEnc.fit_transform(dummy_data[['Only child']])
dummy_data[oneHotEnc.categories_[0]] = res.toarray()
dummy_data.drop('Only child', axis=1, inplace=True)

res = oneHotEnc.fit_transform(dummy_data[['Village - town']])
dummy_data[oneHotEnc.categories_[0]] = res.toarray()
dummy_data.drop('Village - town', axis=1, inplace=True)

res = oneHotEnc.fit_transform(dummy_data[['House - block of flats']])
dummy_data[oneHotEnc.categories_[0]] = res.toarray()
dummy_data.drop('House - block of flats', axis=1, inplace=True)
```
